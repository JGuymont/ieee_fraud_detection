```python
categorical_features = ['ProductCD', 'addr1', 'addr2', 'P_emaildomain', 'R_emaildomain']
categorical_features += ['card{}'.format(i) for i in range(1, 7)]
categorical_features += ['M{}'.format(i) for i in range(1, 10)]
categorical_features += ['isFraud']

numerical_features = [feature for feature in list(df) if feature not in categorical_features]
integer_features = [feature for feature in numerical_features if df[feature].dtype == numpy.int64]
continuous_features = [feature for feature in numerical_features if df[feature].dtype == numpy.float64]
```

```python
df[integer_features]
```

```python
set(df.dtypes)
```

```python
for feature in categorical_features:
    df[feature] = df[feature].astype('category').cat.codes
```

```python
for feature in integer_features:
    if df[feature].dtype == numpy.int64:
        df[feature] = df[feature].astype('int32')
```

```python
for feature in continuous_features:
    if df[feature].dtype == numpy.float64:
        df[feature] = df[feature].astype('float32')
```

```python
df[categorical_features] = df[categorical_features].apply(lambda x: x.cat.codes)
```
